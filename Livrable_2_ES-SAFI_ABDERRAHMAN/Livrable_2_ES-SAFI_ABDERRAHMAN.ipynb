```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

class GridWorld:
    def __init__(self, size=4, start=(0,0), goal=(3,3)):
        self.size = size
        self.start = start
        self.goal = goal
        self.state = start
        self.actions = {
            0: (-1, 0),  # up
            1: (1, 0),   # down
            2: (0, -1),  # left
            3: (0, 1)    # right
        }

    def reset(self):
        self.state = self.start
        return self.state

    def step(self, action):
        dx, dy = self.actions[action]
        x, y = self.state
        new_x, new_y = x + dx, y + dy

        if 0 <= new_x < self.size and 0 <= new_y < self.size:
            self.state = (new_x, new_y)

        if self.state == self.goal:
            return self.state, 1.0, True
        else:
            return self.state, -0.01, False

    def render(self):
        fig, ax = plt.subplots(figsize=(5,5))
        for i in range(self.size):
            for j in range(self.size):
                rect = patches.Rectangle((j, self.size-1-i), 1, 1, linewidth=1, edgecolor='black', facecolor='white')
                ax.add_patch(rect)

        sx, sy = self.start
        gx, gy = self.goal
        ax.add_patch(patches.Rectangle((sy, self.size-1-sx), 1, 1, facecolor='green'))
        ax.add_patch(patches.Rectangle((gy, self.size-1-gx), 1, 1, facecolor='red'))
        ax.add_patch(patches.Rectangle((self.state[1], self.size-1-self.state[0]), 1, 1, facecolor='blue'))

        ax.set_xlim(0, self.size)
        ax.set_ylim(0, self.size)
        ax.set_xticks(range(self.size+1))
        ax.set_yticks(range(self.size+1))
        ax.set_aspect('equal')
        plt.grid(True)
        plt.show(block=False)
        plt.pause(0.3)
        plt.clf()


# ---------------- VALUE ITERATION ----------------
def value_iteration(env, gamma=0.9, theta=1e-4):
    V = np.zeros((env.size, env.size))
    policy = np.zeros((env.size, env.size), dtype=int)

    while True:
        delta = 0
        for x in range(env.size):
            for y in range(env.size):
                if (x, y) == env.goal:
                    continue
                v = V[x, y]
                values = []
                for a in range(4):
                    dx, dy = env.actions[a]
                    nx, ny = x + dx, y + dy
                    if 0 <= nx < env.size and 0 <= ny < env.size:
                        reward = 1.0 if (nx, ny) == env.goal else -0.01
                        values.append(reward + gamma * V[nx, ny])
                    else:
                        # stay in place if move is invalid
                        values.append(-0.01 + gamma * V[x, y])
                V[x, y] = max(values)
                policy[x, y] = np.argmax(values)
                delta = max(delta, abs(v - V[x, y]))
        if delta < theta:
            break
    return V, policy

# ---------------- ANIMATE OPTIMAL POLICY ----------------
def run_policy(env, policy):
    state = env.reset()
    done = False
    while not done:
        env.render()
        action = policy[state]
        state, reward, done = env.step(action)
    plt.close()

# ---------------- MAIN ----------------
env = GridWorld()
V, policy = value_iteration(env)
print("Optimal Policy (0=up,1=down,2=left,3=right):")
print(policy)

run_policy(env, policy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# ---------------- GRIDWORLD ----------------
class GridWorld:
    def __init__(self, size=4, start=(0,0), goal=(3,3)):
        self.size = size
        self.start = start
        self.goal = goal
        self.state = start
        self.actions = {
            0: (-1, 0),  # up
            1: (1, 0),   # down
            2: (0, -1),  # left
            3: (0, 1)    # right
        }

    def reset(self):
        self.state = self.start
        return self.state

    def step(self, action):
        dx, dy = self.actions[action]
        x, y = self.state
        new_x, new_y = x + dx, y + dy

        if 0 <= new_x < self.size and 0 <= new_y < self.size:
            self.state = (new_x, new_y)

        if self.state == self.goal:
            return self.state, 1.0, True
        else:
            return self.state, -0.01, False

    def render(self):
        fig, ax = plt.subplots(figsize=(5,5))
        for i in range(self.size):
            for j in range(self.size):
                rect = patches.Rectangle((j, self.size-1-i), 1, 1, linewidth=1, edgecolor='black', facecolor='white')
                ax.add_patch(rect)

        sx, sy = self.start
        gx, gy = self.goal
        ax.add_patch(patches.Rectangle((sy, self.size-1-sx), 1, 1, facecolor='green'))
        ax.add_patch(patches.Rectangle((gy, self.size-1-gx), 1, 1, facecolor='red'))
        ax.add_patch(patches.Rectangle((self.state[1], self.size-1-self.state[0]), 1, 1, facecolor='blue'))

        ax.set_xlim(0, self.size)
        ax.set_ylim(0, self.size)
        ax.set_xticks(range(self.size+1))
        ax.set_yticks(range(self.size+1))
        ax.set_aspect('equal')
        plt.grid(True)
        plt.show(block=False)
        plt.pause(0.3)
        plt.clf()

# ---------------- POLICY ITERATION ----------------
def policy_evaluation(env, policy, gamma=0.9, theta=1e-4):
    V = np.zeros((env.size, env.size))
    while True:
        delta = 0
        for x in range(env.size):
            for y in range(env.size):
                if (x, y) == env.goal:
                    continue
                a = policy[x, y]
                dx, dy = env.actions[a]
                nx, ny = x + dx, y + dy
                if 0 <= nx < env.size and 0 <= ny < env.size:
                    reward = 1.0 if (nx, ny) == env.goal else -0.01
                    v_new = reward + gamma * V[nx, ny]
                else:
                    v_new = -0.01 + gamma * V[x, y]
                delta = max(delta, abs(V[x, y] - v_new))
                V[x, y] = v_new
        if delta < theta:
            break
    return V

def policy_iteration(env, gamma=0.9):
    # Initialize random policy
    policy = np.random.choice([0,1,2,3], size=(env.size, env.size))
    stable = False
    while not stable:
        V = policy_evaluation(env, policy, gamma)
        stable = True
        for x in range(env.size):
            for y in range(env.size):
                if (x, y) == env.goal:
                    continue
                old_action = policy[x, y]
                action_values = []
                for a in range(4):
                    dx, dy = env.actions[a]
                    nx, ny = x + dx, y + dy
                    if 0 <= nx < env.size and 0 <= ny < env.size:
                        reward = 1.0 if (nx, ny) == env.goal else -0.01
                        action_values.append(reward + gamma * V[nx, ny])
                    else:
                        action_values.append(-0.01 + gamma * V[x, y])
                best_action = np.argmax(action_values)
                policy[x, y] = best_action
                if old_action != best_action:
                    stable = False
    return V, policy

# ---------------- RUN POLICY ----------------
def run_policy(env, policy):
    state = env.reset()
    done = False
    while not done:
        env.render()
        action = policy[state]
        state, reward, done = env.step(action)
    plt.close()

# ---------------- MAIN ----------------
env = GridWorld()
V, policy = policy_iteration(env)
print("Optimal Policy (0=up,1=down,2=left,3=right):")
print(policy)

run_policy(env, policy)
```

# MC

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import random

# ---------------- GRIDWORLD ----------------
class GridWorld:
    def __init__(self, size=4, start=(0,0), goal=(3,3)):
        self.size = size
        self.start = start
        self.goal = goal
        self.state = start
        self.actions = {
            0: (-1, 0),  # up
            1: (1, 0),   # down
            2: (0, -1),  # left
            3: (0, 1)    # right
        }

    def reset(self):
        self.state = self.start
        return self.state

    def step(self, action):
        dx, dy = self.actions[action]
        x, y = self.state
        new_x, new_y = x + dx, y + dy

        if 0 <= new_x < self.size and 0 <= new_y < self.size:
            self.state = (new_x, new_y)

        if self.state == self.goal:
            return self.state, 1.0, True
        else:
            return self.state, -0.01, False

    def render(self):
        fig, ax = plt.subplots(figsize=(5,5))
        for i in range(self.size):
            for j in range(self.size):
                rect = patches.Rectangle((j, self.size-1-i), 1, 1, linewidth=1, edgecolor='black', facecolor='white')
                ax.add_patch(rect)

        sx, sy = self.start
        gx, gy = self.goal
        ax.add_patch(patches.Rectangle((sy, self.size-1-sx), 1, 1, facecolor='green'))
        ax.add_patch(patches.Rectangle((gy, self.size-1-gx), 1, 1, facecolor='red'))
        ax.add_patch(patches.Rectangle((self.state[1], self.size-1-self.state[0]), 1, 1, facecolor='blue'))

        ax.set_xlim(0, self.size)
        ax.set_ylim(0, self.size)
        ax.set_xticks(range(self.size+1))
        ax.set_yticks(range(self.size+1))
        ax.set_aspect('equal')
        plt.grid(True)
        plt.show(block=False)
        plt.pause(0.3)
        plt.clf()

# ---------------- MONTE CARLO CONTROL ----------------
def monte_carlo_control(env, num_episodes=5000, gamma=0.9, epsilon=0.1):
    Q = np.zeros((env.size, env.size, 4))  # state-action values
    returns = [[[] for _ in range(4)] for _ in range(env.size*env.size)]

    def state_to_index(state):
        return state[0]*env.size + state[1]

    for episode in range(num_episodes):
        episode_states = []
        episode_actions = []
        episode_rewards = []

        state = env.reset()
        done = False

        while not done:
            if random.random() < epsilon:
                action = np.random.choice([0,1,2,3])
            else:
                action = np.argmax(Q[state[0], state[1]])

            next_state, reward, done = env.step(action)

            episode_states.append(state)
            episode_actions.append(action)
            episode_rewards.append(reward)

            state = next_state

        G = 0
        visited = set()
        for t in reversed(range(len(episode_states))):
            s = episode_states[t]
            a = episode_actions[t]
            r = episode_rewards[t]
            G = gamma * G + r
            if (s, a) not in visited:
                idx = state_to_index(s)
                returns[idx][a].append(G)
                Q[s[0], s[1], a] = np.mean(returns[idx][a])
                visited.add((s, a))

    policy = np.argmax(Q, axis=2)
    return Q, policy

# ---------------- RUN POLICY WITH VISUALIZATION ----------------
def run_policy(env, policy):
    state = env.reset()
    done = False
    while not done:
        env.render()
        action = policy[state[0], state[1]]
        state, reward, done = env.step(action)
    env.render()
    print(f"Reached goal at {state}!")

# ---------------- MAIN ----------------
env = GridWorld()
Q, policy = monte_carlo_control(env, num_episodes=10000, gamma=0.9, epsilon=0.1)
print("Optimal Policy (0=up,1=down,2=left,3=right):")
print(policy)

run_policy(env, policy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import random

# ---------------- GRIDWORLD WITH PATCHES ----------------
class GridWorld:
    def __init__(self, size=4, start=(0,0), goal=(3,3)):
        self.size = size
        self.start = start
        self.goal = goal
        self.state = start
        self.actions = {
            0: (-1, 0),  # up
            1: (1, 0),   # down
            2: (0, -1),  # left
            3: (0, 1)    # right
        }

    def reset(self):
        self.state = self.start
        return self.state

    def step(self, action):
        dx, dy = self.actions[action]
        x, y = self.state
        new_x, new_y = x + dx, y + dy

        if 0 <= new_x < self.size and 0 <= new_y < self.size:
            self.state = (new_x, new_y)

        if self.state == self.goal:
            return self.state, 1.0, True
        else:
            return self.state, -0.01, False

    def render(self):
        fig, ax = plt.subplots(figsize=(5,5))
        for i in range(self.size):
            for j in range(self.size):
                rect = patches.Rectangle((j, self.size-1-i), 1, 1, linewidth=1, edgecolor='black', facecolor='white')
                ax.add_patch(rect)

        sx, sy = self.start
        gx, gy = self.goal
        ax.add_patch(patches.Rectangle((sy, self.size-1-sx), 1, 1, facecolor='green'))
        ax.add_patch(patches.Rectangle((gy, self.size-1-gx), 1, 1, facecolor='red'))
        ax.add_patch(patches.Rectangle((self.state[1], self.size-1-self.state[0]), 1, 1, facecolor='blue'))

        ax.set_xlim(0, self.size)
        ax.set_ylim(0, self.size)
        ax.set_xticks(range(self.size+1))
        ax.set_yticks(range(self.size+1))
        ax.set_aspect('equal')
        plt.grid(True)
        plt.show(block=False)
        plt.pause(0.3)
        plt.clf()

# ---------------- Q-LEARNING ----------------
def q_learning(env, num_episodes=5000, alpha=0.1, gamma=0.9, epsilon=0.1):
    Q = np.zeros((env.size, env.size, 4))  # state-action values

    for episode in range(num_episodes):
        state = env.reset()
        done = False

        while not done:
            # -greedy action selection
            if random.random() < epsilon:
                action = np.random.choice([0,1,2,3])
            else:
                action = np.argmax(Q[state[0], state[1]])

            next_state, reward, done = env.step(action)

            # Q-Learning update
            best_next_action = np.argmax(Q[next_state[0], next_state[1]])
            Q[state[0], state[1], action] += alpha * (reward + gamma * Q[next_state[0], next_state[1], best_next_action] - Q[state[0], state[1], action])

            state = next_state

    # Derive policy
    policy = np.argmax(Q, axis=2)
    return Q, policy

# ---------------- RUN POLICY VISUAL ----------------
def run_policy_visual(env, policy):
    state = env.reset()
    done = False
    while not done:
        env.render()
        action = policy[state[0], state[1]]
        state, reward, done = env.step(action)
        print(f"Agent at {state}, reward: {reward}")
    # final position
    env.render()
    plt.close()

# ---------------- MAIN ----------------
env = GridWorld()
Q, policy = q_learning(env, num_episodes=10000, alpha=0.1, gamma=0.9, epsilon=0.1)
print("Optimal Policy (0=up,1=down,2=left,3=right):")
print(policy)

run_policy_visual(env, policy)
```
